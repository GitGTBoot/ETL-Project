```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
# CSV/Json files 
us_file='youtube-csv-json/USvideos.csv'
us_json='youtube-csv-json/US_category_id.json'
gb_file='youtube-csv-json/GBvideos.csv'
gb_json='youtube-csv-json/GB_category_id.json'
ca_file ="youtube-csv-json/CAvideos.csv"
cajson_file ="youtube-csv-json/CA_category_id.json"
de_file = ("youtube-csv-json/DEvideos.csv")
dejson_file = ("youtube-csv-json/DE_category_id.json")
```

```python
# reading csv file to pandas df
us_df=pd.read_csv(us_file)
gb_df=pd.read_csv(gb_file)
ca_csv_df = pd.read_csv(ca_file)
de_csv_df = pd.read_csv(de_file)
```

```python
#  reading json file to pandas df
us_json_df=pd.read_json(us_json)
gb_json_df=pd.read_json(gb_json)
cajson_df = pd.read_json(cajson_file)
dejson_df = pd.read_json(dejson_file)
```

```python
us_json_df.head(2)
```

```python
#data cleaning: pulling the title and id from the json df
us_details=[]

for item in us_json_df["items"]:
    title= item['snippet']['title']
    id=int(item['id'])
    us_details.append({'title':title, 'id':id})
    
gb_details=[]

for item in gb_json_df["items"]:
    title= item['snippet']['title']
    id=int(item['id'])
    gb_details.append({'title':title, 'id':id})
    
de_details=[]
for item in dejson_df["items"]:
    title = item["snippet"]["title"]
    id = int(item["id"])
    de_details.append({"title":title, "id":id})
    
ca_details=[]
for item in cajson_df["items"]:
    title = item["snippet"]["title"]
    id = int(item["id"])
    ca_details.append({"title":title, "id":id})
```

```python
us_details[0:5]
```

```python
#loading cleaned json data to pandas df
us_json_df_clean=pd.DataFrame(us_details)
us_json_df_clean.rename(columns = {"id":"category_id","title":"category_title"},inplace=True)

gb_json_df_clean=pd.DataFrame(gb_details)
gb_json_df_clean.rename(columns = {"id":"category_id","title":"category_title"},inplace=True)

ca_json_df_clean=pd.DataFrame(ca_details)
ca_json_df_clean.rename(columns = {"id":"category_id","title":"category_title"},inplace=True)

de_json_df_clean=pd.DataFrame(de_details)
de_json_df_clean.rename(columns = {"id":"category_id","title":"category_title"},inplace=True)
```

```python
us_json_df_clean.head(2)
```

```python
#data cleaning: changing a column to int
us_df.category_id.astype(int)
gb_df.category_id.astype(int)
ca_csv_df.category_id.astype(int)
de_csv_df.category_id.astype(int)
```

```python
# data cleaning: pulling the publish hour from the publish_time column
us_publish_hour = []
for t in us_df["publish_time"]:
    time = t[11:19]
    us_publish_hour.append(time)
    
gb_publish_hour = []
for t in gb_df["publish_time"]:
    time = t[11:19]
    gb_publish_hour.append(time)
    
ca_publish_hour = []
for t in ca_csv_df["publish_time"]:
    time = t[11:19]
    ca_publish_hour.append(time)
    
de_publish_hour = []
for t in de_csv_df["publish_time"]:
    time = t[11:19]
    de_publish_hour.append(time)
```

```python
# adding the cleaned hour information into a column
us_df["publish_hour"] = us_publish_hour
gb_df["publish_hour"] = gb_publish_hour
ca_csv_df["publish_hour"] = ca_publish_hour
de_csv_df["publish_hour"] = de_publish_hour
```

```python
# merging the cdv df with the cleaned json df
us_merged_df = pd.merge(us_df ,us_json_df_clean , how = "left", on = ["category_id","category_id"])
gb_merged_df = pd.merge(gb_df ,gb_json_df_clean , how = "left", on = ["category_id","category_id"])
ca_merged_df = pd.merge(ca_csv_df ,ca_json_df_clean , how = "left", on = ["category_id","category_id"])
de_merged_df = pd.merge(de_csv_df ,de_json_df_clean , how = "left", on = ["category_id","category_id"])
```

```python
#data cleaning: selecting the relevant columns 
us_clean_merged_df = us_merged_df[['video_id','title','channel_title','category_id', 'category_title','trending_date', 
       'publish_hour', 'views', 'likes', 'dislikes']]
us_clean_merged_df.rename(columns={"views":"video_views"}, inplace=True)

gb_clean_merged_df = gb_merged_df[['video_id','title','channel_title','category_id', 'category_title','trending_date', 
       'publish_hour', 'views', 'likes', 'dislikes']]
gb_clean_merged_df.rename(columns={"views":"video_views"}, inplace=True)

ca_clean_merged_df = ca_merged_df[['video_id','title','channel_title','category_id', 'category_title','trending_date', 
       'publish_hour', 'views', 'likes', 'dislikes']]
ca_clean_merged_df.rename(columns={"views":"video_views"}, inplace=True)

de_clean_merged_df = de_merged_df[['video_id','title','channel_title','category_id', 'category_title','trending_date', 
       'publish_hour', 'views', 'likes', 'dislikes']]
de_clean_merged_df.rename(columns={"views":"video_views"}, inplace=True)
```

```python
#database connection
rds_connection_string = "postgres:spitfire1@localhost:5432/youtube_db"
engine = create_engine(f'postgresql://{rds_connection_string}')
```

```python
engine.table_names()
```

```python
# Load DataFrames into postgres SQL database
us_clean_merged_df.to_sql(name='us', con=engine, if_exists='append', index=False)
gb_clean_merged_df.to_sql(name='great_britain', con=engine, if_exists='append', index=False)
ca_clean_merged_df.to_sql(name='canada', con=engine, if_exists='append', index=False)
de_clean_merged_df.to_sql(name='germany', con=engine, if_exists='append', index=False)
```
