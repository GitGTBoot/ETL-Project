```python
import pandas as pd 
from sqlalchemy import create_engine
from datetime import timedelta, datetime, tzinfo
```

```python
cajson_file = "CA_category_id.json"
cajson_df = pd.read_json(cajson_file)
cajson_df.head()
```

```python
cajson_df.iloc[0]["items"].get("id")
```

```python
cajson_df.iloc[0]["items"].get("snippet").get("title")
```

```python
cajson_df.iloc[0]["items"]["snippet"]["title"]
```

```python
ca_details=[]
for item in cajson_df["items"]:
    
    title=(item["snippet"]["title"])
    id = int(item["id"])
    ca_details.append({"title":title, "id":id})
    
print(ca_details)
```

```python
ca_df = pd.DataFrame(ca_details)
ca_df.head()
type(ca_df["id"][0])
```

```python
ca_df_renamed = ca_df.rename(columns = {"id":"category_id","title":"category_title"})
ca_df_renamed.head()
```

```python

# type(ca_df_renamed.category_id.astype(int))
# ca_df_renamed.head()
type(ca_df_renamed["category_id"][0])
```

```python
file =("CAvideos.csv") 
ca_csv_df = pd.read_csv(file)
ca_csv_df.head()
```

```python
 ca_csv_df.category_id.astype(int)
ca_csv_df.head()
```

```python
type(ca_csv_df["category_id"][0])
```

```python
date = ca_csv_df["publish_time"]
date
```

```python
year = int(date[0:4])
month = int(date[5:7])
day = int(date[8:10])
```

```python
publish_hour = []
for t in ca_csv_df["publish_time"]:
    time = t[11:19]
    publish_hour.append(time)
publish_hour
```

```python
ca_csv_df["publish_hour"] = publish_hour
ca_csv_df.head()
```

```python
ca_csv_df.columns
```

```python
ca_csv_df = ca_csv_df[['video_id', 'trending_date', 'title', 'channel_title', 'category_id',
       'publish_time','publish_hour', 'tags', 'views', 'likes', 'dislikes', 'comment_count',
       'thumbnail_link', 'comments_disabled', 'ratings_disabled',
       'video_error_or_removed', 'description', ]]

ca_csv_df.head()
```

```python
merged_df = pd.merge(ca_csv_df , ca_df_renamed, how = "left", on = ["category_id","category_id"])
merged_df
```

```python
merged_df.columns
```

```python
clean_merged_df = merged_df[['video_id','title','channel_title','category_id', 'category_title','trending_date', 
       'publish_hour', 'views', 'likes', 'dislikes']]
clean_merged_df.rename(columns={"views":"video_views"}, inplace=True)
```

```python
clean_merged_df.head()
```

```python
#database connection
rds_connection_string = "postgres:5432@localhost:5432/youtube_db"
engine = create_engine(f'postgresql://{rds_connection_string }')
```

```python
engine.table_names()
```

```python
clean_merged_df.to_sql(name='canada', con=engine, if_exists='append', index=False)
```
